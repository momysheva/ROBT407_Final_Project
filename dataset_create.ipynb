```python
import pandas as pd
import numpy as np
import os
import cv2
from tqdm import tqdm
```

```python
data_path = 'dataset/'
train_data = pd.read_csv(data_path + 'train.csv')
```

```python
labels = train_data['label'].copy()
```

```python
train_data = train_data.drop('label', axis=1)
```

```python
train_data.head()
```

```python
#converting to 2d array of 28x28
train_data_2d = train_data.values.reshape(-1,28,28,)
```

```python
#save each row as a png file on corresponding label folder

#create train folder if not exists
if not os.path.exists('dataset/train'):
    os.makedirs('dataset/train')

for i in tqdm(range(len(train_data_2d))):
    if not os.path.exists('dataset/train/' + str(labels[i])):
        os.makedirs('dataset/train/' + str(labels[i]))
    cv2.imwrite('dataset/train/' + str(labels[i]) + '/' + f"train_{str(i)}" + '.png', train_data_2d[i])
```

```python
test_data = pd.read_csv(data_path + 'test.csv')
test_data.head()
```

```python
#drop id column
test_data = test_data.drop('id', axis=1)
```

```python
test_data_2d = test_data.values.reshape(-1,28,28,)
```

```python
#create test folder if not exists
if not os.path.exists('dataset/test'):
    os.makedirs('dataset/test')

for i in tqdm(range(len(test_data_2d))):
    cv2.imwrite('dataset/test/' + str(i) + '.png', test_data_2d[i])
```
